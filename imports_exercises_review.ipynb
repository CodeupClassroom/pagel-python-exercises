>#### a) How many different ways can you combine a single letter from "abc" with either 1, 2, or 3?
> - This can be visualized as a <a href ='https://www.sqlshack.com/wp-content/uploads/2020/02/sql-cross-join-working-mechanism.png'>SQL Cross Join</a> or a <a href = 'https://miro.medium.com/v2/resize:fit:1400/1*ofgbRmZT2avae5mqvU6c2Q.jpeg'>Cartesian Product</a>

```python
# We want something like:
# a1, a2, a3, b1, b2, ...
```

```python
#what happens if we just use itertools?
#we will get information but not the answer we want

combo = itertools.product("abc",[1,2,3])
print(combo)
```

```python
# returns an object of type `itertools.product`. 
#itertools.product is an **iterator**, so the contents is not output by print()

type(combo)
```

```python
#Option 1
#Convert the itertools.product -> list

list(itertools.product("abc",[1,2,3]))
```

```python
#Option 2
#Return the length of the list

combo = itertools.product("abc",[1,2,3])
print(f"'abc' and 123 can be combined in {len(list(combo))} different ways.")
```

```python
#Option 3
#Unpack the product using List comprehension:

clist = [c for c in itertools.product("abc",[1,2,3])]
print(clist)
```

```python
# or using a `for` loop

for c in itertools.product("abc",[1,2,3]):
    print(c)
```

***NOTE:***
#### 1. Permutation: ***Order matters***. How may possible arrangements can the iterable/input be placed in?
- Think of a **p**assword: `1234` is not the same password as `4321`
>
    
#### 2. Combination: Order ***Doesn't*** matter, **uniqueness matters**. 
- Think of **c**oworkers: `Misty and Andrew` is the same as `Andrew and Misty`

```python
#create a function that makes the desired corrections
#remove $, remove commas, and make a float
def clean_balance(balance):
    '''
    This function will:
        - remove $ and ,
        - convert str to float
    '''
    balance = balance.replace("$", "")
    balance = balance.replace(",", "")
    return float(balance)
```

```python
#check to make sure our function works
clean = clean_balance("$10,213.42")
clean
```

```python
#double check that balance was changed from str to float
#its a float and not int because there are decimal points
type(clean)
```

```python
#use our new function on our ENTIRE DATASET
#for each profile in the profiles
#clean "balance"
for profile in profiles:
    profile['balance'] = clean_balance(profile['balance'])
```

>#### 3.10) Total number of unread messages for all users

```python
#create a variable named messages that takes the greeting from profile and splits it on the space
message = profile['greeting'].split(' ')
```

```python
#call the new variable so we can see what it looks like
message
```

```python
#here we are going to create a for loop with the new message variable we just created
#we are saying that IF the word is a DIGIT
#print "this is a number" and the number from word
for word in message:
    if word.isdigit():
        print('this is a number: ', word)
```

```python
#we are starting with 0
#using what we just learned above
#splitting the greeting on spaces
#if there is a digit in the greeting, we turn that into an integer
#return the total number of unread messages for all users

sum_messages = 0

for profile in profiles:
    message = profile['greeting'].split(' ')
    for word in message:
        if word.isdigit():
            sum_messages = sum_messages + int(word)
            
sum_messages
```
